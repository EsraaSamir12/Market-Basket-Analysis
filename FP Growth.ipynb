```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import fpgrowth, association_rules

dataset = pd.read_csv("D:\\Downloads\\market.csv", header=0, sep=';')
dataset.head(5)
```

```python
# Transform the dataset to a one-hot encoded format
dataset_transformed = dataset.apply(lambda col: col.replace(1, col.name).replace(0, np.nan))
dataset_transformed.columns = range(dataset_transformed.shape[1])
dataset_transformed
```

```python
records = []
for i in range(0, dataset_transformed.shape[0]):
    records.append([str(dataset_transformed.values[i, j]) for j in range(0, dataset_transformed.shape[1]) if pd.notnull(dataset_transformed.values[i, j])])
```

```python
records = []
for i in range(0, dataset_transformed.shape[0]):
    records.append([str(dataset_transformed.values[i, j]) for j in range(0, dataset_transformed.shape[1]) if pd.notnull(dataset_transformed.values[i, j])])
```

```python
# Create a DataFrame for the one-hot encoded format
from mlxtend.preprocessing import TransactionEncoder

encoder = TransactionEncoder()
onehot = encoder.fit(records).transform(records)
df = pd.DataFrame(onehot, columns=encoder.columns_)
```

```python
FG_growth = fpgrowth(df, min_support=0.05, use_colnames=True)
Result = association_rules(FG_growth, metric="lift", min_threshold=1, num_itemsets=2)
Result.head()
```

```python
Result.sort_values("confidence",ascending=False)
Result.head()
```
